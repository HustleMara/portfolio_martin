# Построение дашбордов в Tableau

# Описание проекта

**TED** (от англ. technology, education, design — «технологии, образование, дизайн») — некоммерческий фонд, который проводит популярные конференции. На них выступают специалисты из разных областей и читают лекции на актуальные социальные, культурные и научные темы.

В разное время на TED-конференциях выступали математик Бенуа Мандельброт, теоретик искусственного интеллекта Марвин Минский, спортсменка Дана Ньяд и основатель Google Ларри Пейдж. В истории TED также были неоднозначные и даже скандальные выступления. Например, в 2010 году на конференции выступил Рэнди Пауэлл с рассказом о псевдонаучной «вихревой математике», а в 2014 году в конференции TEDMED участвовала Элизабет Холмс — основательница печально известного стартапа Theranos.

**Описание данных**
* tableau_project_data_1.csv
* tableau_project_data_2.csv
* tableau_project_data_3.csv
* tableau_project_event_dict.csv
* tableau_project_speakers_dict.csv

Файлы tableau_project_data_1.csv, tableau_project_data_2.csv, tableau_project_data_3.csv хранят данные выступлений. 

У них одинаковая структура:

**talk_id** — идентификатор выступления;

**url** — ссылка на запись выступления;

**title** — название выступления;

**description** — краткое описание;

**film_date** — дата записи выступления;

**duration** — длительность в секундах;

**views** — количество просмотров;

**main_tag** — основная категория, к которой относится выступление;

**speaker_id** — уникальный идентификатор автора выступления;

**laughter_count** — количество раз, когда аудитория смеялась в ходе выступления;

**applause_count** — количество раз, когда аудитория аплодировала в ходе выступления;

**language** — язык, на котором велось выступление;

**event_id** — уникальный идентификатор конференции.

**Файл tableau_project_event_dict.csv** — справочник конференций. Описание таблицы:

**conf_id** — уникальный идентификатор конференции;

**event** — название конференции;

**country** — страна проведения конференции.

Файл **tableau_project_speakers_dict.csv** — справочник авторов выступления. Описание таблицы:

**author_id** — уникальный идентификатор автора выступления;

**speaker_name** — имя автора;

**speaker_occupation** — профессиональная область автора;

**speaker_description** — описание профессиональной деятельности автора.

# Ключевые выводы

Лидером по количеству выступлений являются США, на долю которых приходится 50,52% всех выступлений. Следом идут Канада с 19,44% и Великобритания с 8,43%. В других странах, где также проводились выступления, этот показатель варьируется от 0,03% до 1,77%.

За весь период с начала 2009 года до конца июня 2021 года наиболее популярными темами были наука, технологии и культура. В период с 2020 года до середины 2021 года самыми популярными темами стали общество, социальные изменения, культура и глобальные вопросы.

Темы выступлений на темы науки и технологий в Канаде пользуются большей популярностью по сравнению с США и Великобританией. В то же время, тема культуры более востребована в США и Великобритании, чем в Канаде. Великобритания отличается от других стран в топ-3 тем, что в ней проводилось больше выступлений на темы глобальных вопросов, бизнеса и дизайна.

Выступления на темы насилия, здравоохранения, инженерии, робототехники и развлечений в среднем чаще всего вызывают аплодисменты. Реже всего аплодисменты вызывают выступления на темы океана, кино, нейронауки и писательства. Больше всего по отношению к длительности выступления аплодировали автору Сесиль Ричардс с её выступлением на тему социальных изменений под названием "The political progress women have made -- and what's next".

Смех аудитории чаще всего вызывают выступления на темы философии, математики, детей и сообщества, реже всего — на темы работы, насилия, TED Fellows и космоса. Самым смешным оказалось выступление автора Рори Бремнера на тему социальных проблем под названием "A one-man world summit".

Зависимость между длительностью выступления и количеством просмотров отсутствует. Чаще всего смотрели выступление "Your body language may shape who you are", набравшее 63,588,711 просмотров. Самым длительным является выступление "3 secrets to Netflix's success", продолжительностью 63 минуты.

Наиболее распространенные области деятельности авторов включают писательство, журналистику, искусство и предпринимательство. Эти сферы доминируют среди профессиональных направлений, представленных на выступлениях. Обычно у одного автора есть одно выступление, но есть и исключения, когда авторы выступают неоднократно. Например, Ханс Рослинг выступал восемь раз, а Марко Темпест — семь раз. Эти авторы привлекли значительное внимание благодаря своим многократным выступлениям.

Среди дизайнеров чаще всего выступал Том Вуйек, который провёл четыре выступления. Однако его выступление на тему "Learn to use the 13th-century astrolabe" оказалось наименее популярным, набрав 940 542 просмотра.

Темы "science" и "others" вызывают наибольший интерес и смех, что может указывать на их популярность и привлекательность для аудитории. Выступления на темы "technology" и "culture" также популярны, но вызывают меньше смеха. В зависимости от страны, представленные темы могут сильно варьироваться, что свидетельствует о культурных и региональных различиях в интересах аудитории.
